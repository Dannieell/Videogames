```python
# Se realiza el proceso de ETL para ello se importa las librerías json y pandas#
import json
import pandas as pd
```

```python
# Se ubica la ruta del archivo "output_steam_games.json" y se convierte en un dataframe
ruta_archivo_json = "..\Archivos Json\output_steam_games.json"
steam_games  = pd.read_json(ruta_archivo_json, lines=True)
```

```python
# Se verifica si se realizo correctamente la conversion a dataframe
steam_games
```

```python
# Se realiza la conversión de listas a texto
steam_games = steam_games.applymap(lambda x: ', '.join(x) if isinstance(x, list) else x)
```

```python
# Se hace la verificacion viendo las 5 últimas columnas
steam_games.tail(5)
```

```python
# Se hace una revisión rápida de la clase de datos
steam_games.info()
```

```python
# Se puede ver que tenemos filas enteras con valores nulos por ello realiza la depuración de los mismos
steam_games.head()
```

```python
# Eliminar filas en las cuales todos los valores sean nulos o None
steam_games = steam_games.dropna(how='all')

# Reestablecer el índice después de eliminar filas
steam_games.reset_index(drop=True, inplace=True)
```

```python
# Se verifica si la depuración fue correcta observando las primeras 5 columnas
steam_games.head()
```

```python
# También se revisa las últimas 5 columnas
steam_games.tail()
```

```python
# Nueva verificación del dataframe
steam_games.info()
```

```python
# Se guarda el archivo ya transformado a un formato csv
ruta_guardado_csv = "../Datacsv/steam_games.csv"
steam_games.to_csv(ruta_guardado_csv, index=False)
```
